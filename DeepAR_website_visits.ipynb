# Business Case

## Business Context:
Cloth Clothing is a local Kingston clothing store that sells a variety of luxury clothing brands. They have a single physical store location in downtown Kingston but also have a developed website The website provides users with the ability to create an account to save clothing as well the ability to order clothing online for delivery.\
https://clothclothing.ca/

## Problem:
### Issue:
A major problem currently faced by Cloth Clothing is determining how many servers will be needed to handle the user traffic on their website. This is referring to the minimum requirements to allow the website to continuously function without delay or shutdown.
### Assessment:
The inability to correctly adjust the servers needed can lead to excessive waste of server use or a lack of servers when needed. For example, if the user traffic drastically increases to about 500 simultaneous users but the website infrastructure can only handle 400 at a time then it can cause slow loading times, delayed interactions, transaction failure, etc. All of this reduces the customer’s experience whilst trying to navigate Cloth Clothing’s online store and moreover can lead to a loss of potential profits and customers for Cloth Clothing.
### Solution:
By utilizing a time series forecasting model it will be possible for Cloth Clothing to predict how user traffic will look in the coming days. A prediction could show daily user visits substantially increasing in comparison to the current server capacity. This forewarning would allow Cloth Clothing to respond appropriately with enough time to prepare. Similarly, a prediction could show daily user visits begin to drop signaling that the number of servers can be reduced which can save both time and money on server maintenance and electricity.

## Solution Description:
### Business Impact and Value:
There are variety of positive impacts a deep ar model will have on both the customer’s and business.
### Customer Experience:
For the customer it will ensure that the Cloth Clothing’s website will always run smoothly without delay for most times of the year. Thus all customers will enjoy a seamless experience while shopping on the website. This could boost customer interaction as they will be able to view more available products within a fraction of the time they normally would with a slower website.
Of course in order for the deep ar model to continue to have accurate predictions it will require the website to record the IP address of customers on the site as well have them store cookies to verify if there has been a gap between the last time they visited the website. 
### Business Operation:
A time series forecasting model would enable the Cloth Clothing to more easily anticipate peaks or drops in user visits to the online store. The business’s website developers can rest easy and not stress about unanticipated changes as they can send a query to model a prediction for the coming week. Thus the developers will have ample time to prepare changes to the online stores infrastructure if needed.

## Proof of Concept:
### Customer Experience:

# Data Preparation for DeepAR Model Training

```python
# Define a function to convert a pandas Series to a JSON object for DeepAR
def series_to_json_obj(series):
    """
    Converts a pandas Series to a JSON object for DeepAR.

    Parameters:
    series (pandas.Series): The time series data.

    Returns:
    dict: A JSON object with "start" and "target" keys.
    """
    # DeepAR expects each time series to be a JSON object with "start" and "target" keys
    return {"start": str(series.index[0]), "target": list(series)}

# Convert the 'Date' column to string format for compatibility with DeepAR
df_website['Date'] = df_website['Date'].dt.strftime('%Y-%m-%d')

# Split the df_website DataFrame into training and testing data
# For example, use the first 80% of the data for training and the remaining 20% for testing
split_index = int(len(df_website) * 0.8)
train_data = df_website.iloc[:split_index]
test_data = df_website.iloc[split_index:]

# Convert the training data to JSON format for DeepAR
train_series_json = train_data.set_index('Date')['Unique.Visits'].pipe(series_to_json_obj)

# Convert the test data to JSON format for DeepAR
test_series_json = test_data.set_index('Date')['Unique.Visits'].pipe(series_to_json_obj)

# Save the training data JSON object to a file
with open('train_data.json', 'w') as f:
    json.dump(train_series_json, f)

# Save the test data JSON object to a file
with open('test_data.json', 'w') as f:
    json.dump(test_series_json, f)
```

# SageMaker Session and Environment Setup

```python
# Create a SageMaker session for managing interactions with the SageMaker environment
sagemaker_session = sagemaker.Session()

# Retrieve the execution role for the current SageMaker notebook instance
role = get_execution_role()

# Retrieve the AWS region name from the SageMaker session
region = sagemaker_session.boto_region_name

# Specify the DeepAR container
image_uri = retrieve('forecasting-deepar', region)

# Define the default Amazon S3 bucket for the SageMaker session
bucket = sagemaker_session.default_bucket()

# Define the prefix for the S3 key where data will be stored
prefix = 'sagemaker/onlineretail/basic-prediction'

# Define the path to the local training data file
local_train = 'train_data.json'

# Define the path to the local test data file
local_test = 'test_data.json'

# Upload the local training data file to the specified S3 bucket and prefix, returning the S3 URI
train_uri = sagemaker_session.upload_data(local_train, bucket=bucket, key_prefix=prefix)

# Upload the local test data file to the specified S3 bucket and prefix, returning the S3 URI
test_uri = sagemaker_session.upload_data(local_test, bucket=bucket, key_prefix=prefix)

# Print the execution role ARN to the console
print("Execution Role ARN:", role)

# Print the AWS region name to the console
print("AWS Region:", region)

# Print the S3 URI for the training data to the console
print("Training Data S3 URI:", train_uri)

# Print the S3 URI for the test data to the console
print("Test Data S3 URI:", test_uri)
```

# Estimator Configuration for DeepAR Algorithm Training

```python
from sagemaker.estimator import Estimator

# Define an Estimator for the DeepAR algorithm in SageMaker
deepar_estimator = Estimator(
    sagemaker_session=sagemaker_session,  # Specify the SageMaker session
    image_uri=image_uri,  # Specify the container image URI for the DeepAR algorithm
    role=role,  # Specify the IAM role ARN used by SageMaker
    instance_count=1,  # Specify the number of instances to use for training
    instance_type='ml.m5.large',  # Specify the instance type for training
    output_path='s3://{}/output'.format(bucket),  # Specify the S3 location for saving the model artifacts and output
    base_job_name='onlineretail-deepar-forecasting'  # Specify the base job name for the training job
)

# Set hyperparameters for the DeepAR algorithm
deepar_estimator.set_hyperparameters(
    time_freq='D',  # Specify the time frequency of the data (e.g., 'D' for daily)
    context_length=30,  # Specify the length of the context used for training (e.g., 30 days)
    prediction_length=7,  # Specify the length of the prediction horizon (forecasting 7 days into the future)
    epochs=20,  # Specify the number of training epochs
    early_stopping_patience=10,  # Specify the number of epochs with no improvement before early stopping
    num_dynamic_feat='auto',  # Automatically detect dynamic features (if any)
    cardinality='auto'  # Automatically detect the cardinality of categorical features (if any)
)
```
